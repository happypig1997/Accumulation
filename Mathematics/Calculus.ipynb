```python
x = symbols('x')
```

```python
f = x **3 - 50*x **2 +3
diff1 = diff(f)
diff2 = diff(diff1)
diff1,diff2
```

```python
p1 = plot(f,show=False,legend = True)
p2 = plot(diff1,show=False)
p3 = plot(diff2,show=False)
p1.extend(p2)
p1.extend(p3)
p1.show()
# p4.append(p3[0])
```

```python
def y(x):
    y = x **3 - 50*x **2 +3
    return y
def dydx(x):
    return 3*x**2 - 100 * x
def d2ydx2(x):
    return 6*x -100
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(-10,11,0.01)
fig, ax = plt.subplots()
ax.plot(x,y(x),label = 'f')
ax.plot(x,dydx(x),label='f \'')
ax.plot(x,d2ydx2(x),label='f \'\'')
ax.legend()
```

```python
x = symbols('x')
y = symbols('y')
z = symbols('z')
f = x*y - x**2*z + y**2*z
```

```python
print(diff(diff(f,x),x)+
diff(diff(f,y),y)+
diff(diff(f,z),z))
```
